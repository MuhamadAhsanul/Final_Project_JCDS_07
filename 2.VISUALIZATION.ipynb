```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings("ignore")
sb.set_style('whitegrid')
```

```python
df = pd.read_csv("MotorcycleData_Clean.csv", encoding='latin')
df
```

## __Exploratory Data Analysis__

```python
Pada bagian ini, saya mencoba untuk memunculkan grafik korelasi antara Price dengan kolom-kolom yang
saling mempengaruhi 
```

#### __Check Price Distribution__

```python
plt.figure(figsize=(20, 8))
sb.distplot(df['Price'])
```

##### _Price Distribution's by Car Conditions_

```python
plt.figure(figsize=(20,8))
sb.distplot(df[df['Condition'] == 'New']['Price'], label='New')
sb.distplot(df[df['Condition'] == 'Used']['Price'], label='Used')
plt.legend()
```

```python
df[df['Condition'] == 'Used']['Price'].describe()
```

##### _Price Distribution by Model_

```python
plt.figure(figsize=(20, 8))
sb.distplot(df[df['Model'] == 'Touring']['Price'], label='Touring')
sb.distplot(df[df['Model'] == 'Softail']['Price'], label='Softail')
sb.distplot(df[df['Model'] == 'Dyna']['Price'], label='Dyna')
sb.distplot(df[df['Model'] == 'Sportster']['Price'], label='Sportster')
sb.distplot(df[df['Model'] == 'V-Rod']['Price'], label='V-Rod')
sb.distplot(df[df['Model'] == 'Street Glide']['Price'], label='Street Glide')
sb.distplot(df[df['Model'] == 'Heritage Softail Classic']['Price'], label='Heritage Softail Classic')
sb.distplot(df[df['Model'] == 'Other']['Price'], label='Other')
plt.legend()
```

##### _Price Distribution by Type_

```python
plt.figure(figsize=(20, 8))
sb.distplot(df[df['Type'] == 'Gasoline']['Price'], label='Gasoline')
sb.distplot(df[df['Type'] == 'Touring']['Price'], label='Touring')
sb.distplot(df[df['Type'] == 'Cruiser']['Price'], label='Cruiser')
sb.distplot(df[df['Type'] == 'Standard']['Price'], label='Standard')
sb.distplot(df[df['Type'] == 'Sport Touring']['Price'], label='Sport Touring')
sb.distplot(df[df['Type'] == 'Sport Bike']['Price'], label='Sport Bike')
sb.distplot(df[df['Type'] == 'Dual Purpose']['Price'], label='Dual Purpose')
plt.legend()
```

##### _Price Distribution by Seller Status_

```python
plt.figure(figsize=(20, 8))
sb.distplot(df[df['Seller_Status'] == 'dealer']['Price'], label='Dealer')
sb.distplot(df[df['Seller_Status'] == 'private seller']['Price'], label='Private Seller')
sb.distplot(df[df['Seller_Status'] == 'owner']['Price'], label='Owner')
plt.legend()
```

##### _Price Distribution by Vehicle Tile_

```python
plt.figure(figsize=(20, 8))
sb.distplot(df[df['Vehicle_Tile'] == 'Clear']['Price'], label='Baru')
sb.distplot(df[df['Vehicle_Tile'] == 'Salvage']['Price'], label='Bekas')
sb.distplot(df[df['Vehicle_Tile'] == 'Rebuilt, Rebuildable &amp; Reconstructed']['Price'], label='Rebuilt')
plt.legend()
```

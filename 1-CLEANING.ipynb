```python
df2 = df1.copy()
df2
```

```python
df2.Seller_Status = df2.Seller_Status.apply(lambda x: str(x).lower())
df2
```

### Feature Engineering

```python
type_stats = df2.groupby('Type')['Type'].agg('count').sort_values(ascending=False)
type_stats.head(10)
```

```python
model_stats = df2.groupby('Model')['Model'].agg('count').sort_values(ascending=False)
model_stats.head(10)
```

```python
seller_stats = df2.groupby('Seller_Status')['Seller_Status'].agg('count').sort_values(ascending=False)
seller_stats.head()
```

```python
TSLT_25 = type_stats[type_stats<=25] # Type Stats Less Than 25
TSLT_25
```

```python
MSLT_30 = model_stats[model_stats<=30] # Model Stats Less Than 25
MSLT_30
```

```python
SSLT_50 = seller_stats[seller_stats<=50] # Model Stats Less Than 25
SSLT_50
```

```python
df2.Type = df2.Type.apply(lambda x: 'Other' if x in TSLT_25 else x)
df2.Model = df2.Model.apply(lambda x: 'Other' if x in MSLT_30 else x)
df2.Seller_Status = df2.Seller_Status.apply(lambda x: 'Other' if x in SSLT_50 else x)
df2
```

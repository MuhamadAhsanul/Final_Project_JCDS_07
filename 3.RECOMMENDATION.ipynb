```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

<hr>

### Import Dataset

```python
df = pd.read_csv("MotorcycleData_Clean.csv")
df.head(10)
```

### - Parameter 1 -
Based on Price, with 3 Parameter. Default:
- Condition
- Model
- Type

```python
from sklearn.feature_extraction.text import CountVectorizer
df['par1'] = df['Condition'] + ',' + df['Model'] + ',' + df['Type'] 

ext = CountVectorizer(
        tokenizer = lambda x: x.split(';'))
param1 = ext.fit_transform(df['par1'])

print(ext.get_feature_names()) #tokenizer
print(len(ext.get_feature_names()))
df
```

```python
score1 = cosine_similarity(param1)
score1
```

### - Parameter 2 -
Without Price, 4 parameter:
- Condition
- Model
- Type
- Vehicle Tile

```python
from sklearn.feature_extraction.text import CountVectorizer
df['par2'] = df['Condition'] + ';' + df['Model'] + ',' + df['Type'] + ',' + df['Vehicle_Tile']

ext1 = CountVectorizer(
        tokenizer = lambda x: x.split(';'))
param2 = ext1.fit_transform(df['par2'])

print(ext1.get_feature_names()) #tokenizer
print(len(ext1.get_feature_names()))
df
```

```python
score2 = cosine_similarity(param2)
score2
```

### - Parameter 3 -
Without Price, 4 parameter:
- Condition
- Model
- Type
- Seller Status

```python
from sklearn.feature_extraction.text import CountVectorizer
df['par3'] = df['Condition'] + ';' + df['Model'] + ',' + df['Type'] + ',' + df['Seller_Status']

ext3 =  CountVectorizer()
param3 = ext3.fit_transform(df['par3'])

print(ext3.get_feature_names())
print(len(ext3.get_feature_names()))
df
```

```python
score3 = cosine_similarity(param3)
score3
```
